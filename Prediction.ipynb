```python
import pandas as pd
col_names=['sno','Gender','Age','Symptom1','Symptom2','Symptom3','Symptom4','Symptom5','Symptom6','Symptom7','Disease']
df=pd.read_csv('patient.csv',names=col_names)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df.Disease=le.fit_transform(df.Disease)
df.head()
```

```python
x=df.Age
y=df.Disease
plt.pie(y,labels=x)
```

```python
x=df['prognosis']
```

```python
x.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
xpos=np.arange(len(x))
xpos
```

```python
fig=plt.figure(figsize=(24,24))
plt.plot(xpos,y)
plt.xticks(xpos,x)
plt.xticks(rotation=90)
plt.xlabel('Diseases',fontsize=15)
plt.ylabel('Frequency of symptoms',fontsize=15)
plt.title('Prediction of Symptoms ',fontsize=15)
plt.legend()
plt.show()
```
